```python
df2_ds['EdLevel'].unique()
```

```python
for edlevel in df2_ds['EdLevel'].dropna().unique():
    value = df2_ds[df2_ds['EdLevel'] == edlevel]['EdLevel'].count()
    result = round(value/total*100, 2)
    print('{}: {}% study data science'.format(edlevel, result))
    
# value_counts() method does this same thing!
```
